## Program:-1 Write a program that prompts the user to enter the number of students and each student’s name and score and display the student name and highest score**

```python
# Ask the user how many students there are
num_students = int(input("Enter the number of students: "))

# Create empty lists to store names and scores
names = []
scores = []

# Loop to get each student's name and score
for i in range(num_students):
    name = input(f"Enter the name of student {i + 1}: ")
    score = float(input(f"Enter the score of {name}: "))
    names.append(name)
    scores.append(score)

# Find the highest score and the student's name
highest_score = max(scores)
index = scores.index(highest_score)

# Display the result
print("\nThe student with the highest score is:", names[index])
print("Highest score:", highest_score)
```

- In this program 1 of week 2, the program first asks the user how many students’ data will be entered by the users.
- Input() to get data and print() to display results.
- It first asks for the number of students, then uses a loop to collect each student’s name and score and how many times the loop will run.
- Two lists are created to store the names and scores of students.
- For i in range(num_students) is used to repeatedly take input for each student and inside the loop, the program asks for the student’s name
  and score. 
- Both are added to their lists using the .append() method.
- Using the max() function, the program finds the highest score.
- Scores.index(highest_score) returns the position of that highest score.
- .append(), max(), and .index() are used to manage and search data.

## Program:-1 Ask user for an input of his/her password and store it in a variable called password.Print the password length first then create function validating if the password has more than eight characters. If the password fulfills the above condition and has more then eight character  print 'valid password' otherwise print invalid password and ask the user to enter the password again.

```python
# Ask the user to enter a password and store it in a variable
user_password = input("[+] Your password: ")

# Print the length of the entered password using the len() function
print(f"[+] Your password length: {len(user_password)}")

# Define a function to check if the password is valid
# Here, the password is considered valid if its length is at least 8 characters
def is_password_valid(password: str) -> bool:
    # Returns True if length >= 8, otherwise False
    return True if len(password) >= 8 else False 

    # Check if password contains at least one digit using regex
    if not re.search(r"\d", password):  # <-- Regular Expression used here
        return False
    # Check if password contains at least one special character using regex
    if not re.search(r"[!@#$%^&*(),.?\":{}|<>]", password):  # <-- Regular Expression used here
        return False
    return True


# Keep asking the user to enter a password until it is valid
while not is_password_valid(user_password):
    # Prompt the user again if the password is too short
    user_password = input("[-] Your password is invalid, enter it again: ")

# If the loop ends, the password is valid
print("[+] Valid Password!")
```

## Program 1:-  Simple Shopping List
               Create a program that:

               Writes a shopping list to "shopping.txt" (5 items, one per line)
               Reads the file and prints "You need to buy: [item]" for each item Counts and displays the total number of items


```python

# Define shopping list

shopping_items = ["Milk", "Bread", "Eggs", "Apples", "Cheese"]

# Store items in different data structures
shopping_tuple = tuple(shopping_items)       # Tuple: ordered and immutable
shopping_set = set(shopping_items)           # Set: unique, unordered
shopping_dict = {item: 1 for item in shopping_items}  # Dictionary: default quantity 1

# Write shopping list to file

try:
    with open("shopping.txt", "w") as file:
        for item in shopping_items:
            file.write(item + "\n")
    print("[+] Shopping list saved to 'shopping.txt'.")
except Exception as e:
    print(f"Error writing to file: {e}")

# Read shopping list from file

try:
    with open("shopping.txt", "r") as file:
        items_from_file = [line.strip() for line in file.readlines()]  # Remove newline characters
except FileNotFoundError:
    print("The file 'shopping.txt' was not found.")
    items_from_file = []
except Exception as e:
    print(f"Error reading file: {e}")
    items_from_file = []

print("\nShopping List from file:")                              # Display items from file
for item in items_from_file:
    print("You need to buy:", item)

print("\nTotal number of items in list:", len(items_from_file))   # Count total items

# Display data structures

print("\nTuple of items:", shopping_tuple)
print("Set of items:", shopping_set)

print("\nDictionary of items with quantity:")
for item, qty in shopping_dict.items():
    print(f"{item}: {qty}")

# Pandas DataFrame representation
df = ({
    "Item": list(shopping_dict.keys()),
    "Quantity": list(shopping_dict.values())
})
print("\nPandas DataFrame:")
print(df)
```
